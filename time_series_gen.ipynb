```python
from statsmodels.tsa import arima_process
```

# Generate tabular data for 2 EHRs

```python
import numpy as np
from sklearn import datasets
arparams = np.array([.75, -.25])
maparams = np.array([.65, .35])
ar = np.r_[1, -arparams] # add zero-lag and negate
ma = np.r_[1, maparams] # add zero-lag
mean = np.zeros(20)
cov = datasets.make_spd_matrix(20) # 20 temporal generated features
rand_gen = lambda size: np.random.multivariate_normal(mean, cov, size)
```

```python
a = arima_process.arma_generate_sample(ar, ma, nsample=(5000,24), distrvs=rand_gen,  axis = 1) # 1000: num_of_patients, 24: time stamps(axis=1), 20: num_temporal_features
a.shape
```

### check the correlation between different temporal featurs in generated data

```python
reshape_a = a.reshape([-1, 20])
reshape_a_df = pd.DataFrame(reshape_a, columns=["f"+ str(i) for i in range(1, 21)])
reshape_a_df.corr()
```

# Devide a(including 2 EHR's generated temporal data) into 2 EHR data

```python
gen_EHR1 = np.split(a, 2, axis=0)[0]
gen_EHR2 = np.split(a, 2, axis=0)[1]
print(gen_EHR1.shape)
print(gen_EHR2.shape)
```

```python
import pandas as pd
gen_EHR1_data = gen_EHR1.reshape([-1, 20])
gen_EHR1_df = pd.DataFrame(gen_EHR1_data, columns=["f"+ str(i) + "_1" for i in range(1, 21)])
gen_EHR1_df.head()
```

```python
gen_EHR2_data = gen_EHR2.reshape([-1, 20])
gen_EHR2_df = pd.DataFrame(gen_EHR2_data, columns=["f"+ str(i) + "_2" for i in range(1, 21)])
gen_EHR2_df.head()
```

# Define pre-mapped and unmapped features
* pre-mapped: first 5 columns for both EHR
* unmapped: remaining 15 columns for EHR1, shuffled remaining 15 columns for EHR2

# Keep gen_EHR1_df as it is, shuffled the last 15 columns of gen_EHR2_df as unmapped features

```python
import random
ump_cols_2 = list(gen_EHR2_df.columns[5:])
random.shuffle(ump_cols_2)
new_gen_EHR2_df_cols = list(gen_EHR2_df.columns[:5]) + ump_cols_2
gen_EHR2_df_ump_sfl = gen_EHR2_df[new_gen_EHR2_df_cols]
gen_EHR2_df_ump_sfl.head()
```

```python
gen_EHR1_df.to_csv("./data_EHR1.csv", index=False)
```
