```python
from statsmodels.tsa import arima_process
```

# Generate tabular data for EHR1

```python
import numpy as np
from sklearn import datasets
arparams = np.array([.75, -.25])
maparams = np.array([.65, .35])
ar = np.r_[1, -arparams] # add zero-lag and negate
ma = np.r_[1, maparams] # add zero-lag
mean = np.zeros(20)
cov1 = datasets.make_spd_matrix(20) # 20 temporal generated features
rand_gen1 = lambda size: np.random.multivariate_normal(mean, cov1, size)
```

```python
a = arima_process.arma_generate_sample(ar, ma, nsample=(1000,24), distrvs=rand_gen1,  axis = 1) # 1000: num_of_patients, 24: time stamps(axis=1), 20: num_temporal_features
a.shape
```

```python
import pandas as pd
gen_rec1 = a.reshape([-1, 20])
gen_rec1.shape
```

```python
rec_df_1 = pd.DataFrame(gen_rec1, columns=["f"+ str(i) + "_1" for i in range(1, 21)])
rec_df_1.head()
```

```python
rec_df_1.shape
```

```python
rec_df_1.corr()
```

# Generate tabular data for EHR2

```python
cov2 = datasets.make_spd_matrix(20) # 20 temporal generated features
rand_gen2 = lambda size: np.random.multivariate_normal(mean, cov2, size)
```

```python
b = arima_process.arma_generate_sample(ar, ma, nsample=(1000,24), distrvs=rand_gen2,  axis = 1) # 1000: num_of_patients, 24: time stamps(axis=1), 20: num_temporal_features
b.shape
```

```python
import pandas as pd
gen_rec2 = b.reshape([-1, 20])
rec_df_2 = pd.DataFrame(gen_rec2, columns=["f"+ str(i) + "_2" for i in range(1, 21)])
rec_df_2.head()
```

```python
rec_df_2.shape
```

```python
rec_df_2.corr()
```

# Get correlation between 2 tabulars' columns

```python
corr_btw_2_gen = pd.DataFrame(np.zeros((20, 20)), index=rec_df_1.columns, columns=rec_df_2.columns)
for i in range(1, 21):
    cur_col_2 = "f"+str(i)+"_2"
    corr_btw_2_gen[cur_col_2] = rec_df_1.corrwith(rec_df_2[cur_col_2])
    
corr_btw_2_gen
```
